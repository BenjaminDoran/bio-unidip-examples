```python
def ecdf(x):
    """ Return CDF of array """
    return np.arange(len(x))/float(len(x))
```

```python
def plot_ecdf(a):
    """ Return plot of CDF by sorted xvals """
    import matplotlib.pyplot as plt 
    sorted_ = np.sort(a)
    yvals = np.arange(len(sorted_))/float(len(sorted_))
    return plt.plot(sorted_, yvals)
```
